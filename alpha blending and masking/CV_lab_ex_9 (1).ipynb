# Alpha Blending

```python
import pandas as pd
import numpy as np
import cv2
```

```python
img2 = cv2.resize(img2, (225,225), interpolation = cv2.INTER_AREA)
```

```python
img2.shape
```

```python
alpha = 0.8
```

```python
beta = (1.0 - alpha)
dst = cv2.addWeighted(img1, alpha, img2, beta, 0.0)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(dst)
plt.show()
```

# Alpha Masking

```python
img1 = cv2.imread('sample.jfif')
img2 = cv2.imread('vector.jfif')
```

```python
plt.imshow(img1)
plt.show()
```

```python
plt.imshow(img2)
plt.show()
```

```python
img1.shape
```

```python
img2 = cv2.resize(img2, (225,225), interpolation = cv2.INTER_AREA)
```

```python
dst = cv2.bitwise_and(img1, img2)
```

```python
plt.imshow(dst)
plt.show()
```
